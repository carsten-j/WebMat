## opgave 14

Afstandsformlen for distancen mellem 2 punkter $P=(x_1,y_1)$ og $Q=(x_2, y_2)$ siger, at afstanden mellem $P$ og $Q$ er givet ved $\sqrt{(x_2-x_1)^2 + (y_2-y_1)^2  }$.

Derfor er afstanden mellem punktet $P=(-12,0)$ og et vilkårligt punkt på cirklen $Q\big(x(t),y(t)\big)=\big(5\cos(t), 2+5\sin(t)\big)$ givet ved

$\sqrt{\big(5\cos(t) + 12\big)^2 + \big(2+5\sin(t)\big)^2} =$

$\sqrt{25\cos^2(t) + 144 + 120\cos(t) + 4 + 25\sin^2(t) + 20\sin(t)}=$

$\sqrt{25(\cos^2(t)+ \sin^2(t)) + 144 +4 + 120\cos(t) + 20\sin(t)}=$

$\sqrt{25 + 144 +4 + 120\cos(t) + 20\sin(t)}=$

$\sqrt{173 + 120\cos(t) + 20\sin(t)}$

Hvis man skal finde det punkt på cirklen som har kortest afstand til $P$, så kan man gøre det vil at finde det $t$ som er en minimum for afstandsfunktionen.

```python
t = sp.symbols('t', real=True)
```

```python
d = sp.sqrt(173 + 120 * sp.cos(t) + 20 * sp.sin(t))
d
```

For at finde minimum for $d$ kigger vi på den afledede og finder nulpunkterne for denne.

```python
dt = sp.diff(d, t)
dt
```

Lad os se på grafen for $d$ og $dt$. Vi ser at $dt$ har 2 nulpunkter i intervallet 0 til $2\pi$.

```python
dt_np = sp.lambdify(t, dt, 'numpy')
d_np = sp.lambdify(t, d, 'numpy')
```
