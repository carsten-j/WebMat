```python
from scipy.special import binom as bn
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt 
import plotly.express as px
```

```python
n = 1000
p = 0.5
```

```python
x = range(n+1)
dist = [binom.pmf(k, n, p) for k in x]
```

```python
cdf = [binom.cdf(k, n, p) for k in x]
```

```python
mean, var, skew, kurt = binom.stats(n, p, moments='mvsk')
```

```python
fig = px.bar(x=x, y=dist)
fig.show()
```

```python
students = {'x': [0, 4, 7, 11],
            'y': [1876, 2043, 2136, 2323]}

student_data = pd.DataFrame(data=students)
```

```python
model = np.polyfit(student_data.x,student_data.y, 1)
model
```

```python
m = lambda y: model[0] * y + model[1]
```
