```python
import plotly.express as px
import numpy as np
```

Funktion $f$ har netop to tangenter $t_1$ og $t_2$, der går gennem punktet $(1,-3)$.

Bestem en ligning for disse to tangenter

```python
def f(x):
    return x**2
```

Den afledede til $f$ er $f'(x)=2x$ og derfor bliver den generelle tangentligning

$y = 2x_0(x-x_0) + x_0^2$

hvis denne går gennem punktet $(1, -3)$ så fås

$-3 = 2x_0(1-x_0) + x_0^2$

dvs

$-x_0^2 + 2x_0 + 3 = 0$

som har rødderne -1 og 3. Hvis vi indsætter disse værdier i tangentligningen finder vi de to tangentligninger til

$t_1$:

```python
def g(x):
    return -2*x - 1
```

$t_2$:

```python
def h(x):
    return 6*x - 9
```

```python
x = np.linspace(-4, 6, 100)
```

```python
fig = px.line(x=x, y=[f(x), g(x), h(x)])
fig.update_yaxes(range=[-5, 20])
fig.show()
```
