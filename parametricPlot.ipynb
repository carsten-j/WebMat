I et koordinatsystem i planen er der x(t)=sin(t) og y(t)=t^2-4 i intervallet -4≤t≤4.

Bestem vinklen mellem de to hastighedsvektorer i dobbeltpunktet.

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
from sympy import *    
from sympy.plotting import plot_parametric
import math
import numpy as np
import plotly.express as px
init_printing()        # for nice math output 
```

```python
t = symbols('t')
```

```python
c = (np.pi**2-4)
x = sin(t)
y = t**2 - 4
p1 = plot_parametric(x, y, (t, -4, 4), show=false)
p2 = plot( 2*np.pi*t + c, (t, -1, 0), show=false)
p3 = plot(-2*np.pi*t + c, (t, -1, 0), show=false)
```

```python
p1.extend(p2)
p1.extend(p3)
p1[1].line_color = 'red'
p1[2].line_color = 'green'
```

```python
p1.show()
```

```python
u = np.arange(-4, 4, 0.01)
fig = px.line(x=u, y=np.sin(u))
fig.show()
```

```python
solveset(sin(t), t, Interval(-4, 4))
```

```python
dx = diff(x, t)
dy = diff(y, t)
v = [dx, dy]
```

```python
v1 = np.array([i.subs(t, -np.pi) for i in v]).astype(np.float64)
v1
```

```python
v2 = np.array([i.subs(t, np.pi) for i in v]).astype(np.float64)
v2
```

```python
def vinkel_mellem_vektorer(v1, v2):
    res = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
    return math.degrees(math.acos(res))
```

```python
vinkel_mellem_vektorer(v1, v2)
```
